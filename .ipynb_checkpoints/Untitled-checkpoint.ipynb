```python
#특정 년도들의 평균기온 값
locList = ['서울', '부산', '울산', '제주']
temp = []
years = ['18', '19', '20']
res = []
for i in range(len(years)):
    for loc in locList:
        temp.append(df[(df['지점명'] == loc) & (df['일시'].str.contains(years[i]))])
    line = []
    
    for j in range(len(locList)):
        m = temp[j]['평균기온(°C)'].mean()
        line.append(m)

    d = {
        '지점명': locList,
        '평균기온': line
    }
    res.append(d)

res
```
